## Importing Libraries

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import DataLoader
from torchvision import datasets, transforms
```

## 2. LeNet Architecture (PyTorch)

```python
class LeNet(nn.Module):
    def __init__(self):
        super(LeNet, self).__init__()

        self.conv1 = nn.Conv2d(1, 6, kernel_size=5)        # output: 6×28×28
        self.pool1 = nn.AvgPool2d(kernel_size=2, stride=2) # output: 6×14×14

        self.conv2 = nn.Conv2d(6, 16, kernel_size=5)       # output: 16×10×10
        self.pool2 = nn.AvgPool2d(kernel_size=2, stride=2) # output: 16×5×5

        self.fc1 = nn.Linear(16 * 5 * 5, 120)
        self.fc2 = nn.Linear(120, 84)
        self.fc3 = nn.Linear(84, 10)

    def forward(self, x):
        x = torch.tanh(self.conv1(x))
        x = self.pool1(x)

        x = torch.tanh(self.conv2(x))
        x = self.pool2(x)

        x = x.view(x.size(0), -1)  # flatten

        x = torch.tanh(self.fc1(x))
        x = torch.tanh(self.fc2(x))
        x = self.fc3(x)            # note: softmax is applied in loss
        return x


model = LeNet()
print(model)
```

```python
from torchsummary import summary
summary(model, input_size=(1, 32, 32))
```
